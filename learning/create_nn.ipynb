# 创建神经网络
torch.nn包用于构建网络
nn.Module中有层(layer)，以及一个返回 output 的forward方法。

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

我们接下来定义一个网络，继承自父类nn.Module 使用__init__函数定义神经网络的全连接层。这里的__init__函数初始化了模型内的网络层，但是还没有说明数据如何经过这些网络层。

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()

        self.conv1 = nn.Conv2d(1, 32, 3, 1)
        self.conv2 = nn.Conv2d(32, 64, 3, 1)

        self.dropout1 = nn.Dropout2d(0.25)
        self.dropout2 = nn.Dropout2d(0.50)

        self.fc1 = nn.Linear(9216, 128)
        self.fc2 = nn.Linear(128, 10)

my_nn = Net()
print(my_nn)
        
```

指定数据如何穿过我们实现定义的网络层，即指定数据如何经过我的模型。需要我们定义forward函数，对应着前向传播。

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()

        self.conv1 = nn.Conv2d(1, 32, 3, 1)
        self.conv2 = nn.Conv2d(32, 64, 3, 1)

        self.dropout1 = nn.Dropout2d(0.25)
        self.dropout2 = nn.Dropout2d(0.50)

        self.fc1 = nn.Linear(9216, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.conv1(x)
        x = F.relu(x)

        x = self.conv2(x)
        x = F.relu(x)

        x = F.max_pool2d(x,2)
        x = self.dropout1(x)
        x = torch.flatten(x, 1)

        x = self.fc1(x)
        x = F.relu(x)

        x = self.dropout2(x)
        x = self.fc2(x)

        output = F.log_softmax(x, dim = 1)
        return output
```

我们使用一个随机生成的数据集来测试我们的模型

```python
random_data = torch.rand((1, 1, 28, 28))

my_nn = Net()
result = my_nn(random_data)
print(result)
```
